```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
perps = []
steps = []
for n in ('lmperplexity.err', 'lmtrainperplexity.err'):
    with open('lmperplexity.err') as f:
        text = f.read()
        lines = text.split('\n')
        perps.append([float(l.split(',')[0]) for l in lines if l != ''])
        steps.append([float(l.split(',')[1]) for l in lines if l != ''])
```

```python
plt.plot(steps[0], perps[0])
plt.title('Word+POS LM with learning rate halving')
plt.xlabel('Training round')
plt.ylabel('Marginal (word) validation perplexity')
plt.show()
```

```python
plt.plot(steps[1], perps[1])
plt.title('Word+POS LM with learning rate halving')
plt.xlabel('Training round')
plt.ylabel('Marginal (word) training perplexity')
plt.show()
```
